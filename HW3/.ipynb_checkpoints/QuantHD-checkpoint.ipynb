```python
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm

import QuantHD
```

```python
DATA_SET = 'mnist'
DIM = [5120, 6144, 7168, 8192, 9216, 10240, ]
MAX_EPOCHS = 30
LR = 0.05

mapping_by_bit = {
    1: [0.01258386, 6.48191174],
    2: [0.01258386, 0.93122196, 3.29227301, 6.48191174],
    3: [0.01258386, 0.19029319, 0.69736032, 1.48546665, 2.50629356, 3.71152245, 5.05283471, 6.48191174],
    4: [0.01258386, 0.05145483, 0.16766473, 0.35630295, 0.61245889, 0.93122196, 1.30768154, 1.73692704, 2.21404785, 2.73413338, 3.29227301, 3.88355615, 4.5030722, 5.14591054, 5.80716059, 6.48191174],
}
```

```python
dl = QuantHD.Dataloader(dir=DATA_SET, dataset=DATA_SET, data_loc='./Conventional_Data/')
nFeatures, nClasses, traindata, trainlabels, testdata, testlabels = dl.getParam()
traindata = traindata[:10000]
testdata = testdata[:2000]
```

```python
def full_precision(D, nClasses, trainencoded, trainlabels, testencoded, testlabels, epochs, lr=1.):
    
    history = []
    max_test_score = 0.

    retrain_hdc = QuantHD.QHD_classifier(D, nClasses, 0)
    retrain_hdc.init_train(trainencoded, trainlabels, lr=lr)
    trainscore = retrain_hdc.test(trainencoded, trainlabels)
    testscore = retrain_hdc.test(testencoded, testlabels)

    history.append({'train_acc': trainscore, 'test_acc': testscore})

    for epoch in tqdm(range(epochs)):
        retrain_hdc.itr_train(trainencoded, trainlabels, lr=lr)
        trainscore = retrain_hdc.test(trainencoded, trainlabels)
        testscore = retrain_hdc.test(testencoded, testlabels)

        history.append({'train_acc': trainscore, 'test_acc': testscore})

        max_test_score = max(max_test_score, testscore)
    
    return max_test_score, history

def low_precision(bits, D, nClasses, trainencoded, trainlabels, testencoded, testlabels, epochs, std=0., lr=1.):
    
    history = []
    max_test_score = 0.

    retrain_hdc = QuantHD.QHD_classifier(D, nClasses, bits)
    retrain_hdc.init_train(trainencoded, trainlabels, lr=lr)
    retrain_hdc.model_projection()
    retrain_hdc.gen_pre_sampled(std, mapping_by_bit[bits])
    # flip_rate = retrain_hdc.random_bit_flip(std) # bits error
    trainscore = retrain_hdc.test(trainencoded, trainlabels, std=std, mapping=mapping_by_bit[bits], use_pre_sampled=True)
    testscore = retrain_hdc.test(testencoded, testlabels, std=std, mapping=mapping_by_bit[bits], use_pre_sampled=True)

    history.append({'train_acc': trainscore, 'test_acc': testscore})

    for epoch in tqdm(range(epochs)):
        retrain_hdc.itr_train(trainencoded, trainlabels, lr=lr, std=std, mapping=mapping_by_bit[bits], use_pre_sampled=True)
        retrain_hdc.model_projection()
        retrain_hdc.gen_pre_sampled(std, mapping_by_bit[bits])
        #flip_rate = retrain_hdc.random_bit_flip(std) # bits error
        trainscore = retrain_hdc.test(trainencoded, trainlabels, std=std, mapping=mapping_by_bit[bits], use_pre_sampled=True)
        testscore = retrain_hdc.test(testencoded, testlabels, std=std, mapping=mapping_by_bit[bits], use_pre_sampled=True)

        history.append({'train_acc': trainscore, 'test_acc': testscore})

        max_test_score = max(max_test_score, testscore)
    
    return max_test_score, history
```

```python
for dim in DIM:
    
    hdb = QuantHD.HD_basis(dim, nFeatures)
    basis = hdb.getBasis()

    hde = QuantHD.HD_encoder(basis)
    trainencoded = hde.encodeData(traindata)
    testencoded = hde.encodeData(testdata)

    max_test_acc, full_precision_history = full_precision(dim, nClasses, trainencoded, trainlabels, testencoded, testlabels, MAX_EPOCHS, LR)

    plt.plot([result['train_acc'] for result in full_precision_history], label='train')
    plt.plot([result['test_acc'] for result in full_precision_history], label='test')
    plt.legend()
    plt.title(f'Full precision, D={dim}')
    plt.xlabel('epochs')
    plt.ylabel('accuracy')
    plt.show()
```

```python
for dim in DIM:
    
    hdb = QuantHD.HD_basis(dim, nFeatures)
    basis = hdb.getBasis()

    hde = QuantHD.HD_encoder(basis)
    trainencoded = hde.encodeData(traindata)
    testencoded = hde.encodeData(testdata)

    for bit in [1, 2, 3, 4]:
        
        max_test_acc, low_precision_history = low_precision(bit, dim, nClasses, trainencoded, trainlabels, testencoded, testlabels, MAX_EPOCHS, std=0., lr=LR)

        plt.plot([result['train_acc'] for result in low_precision_history], label='train')
        plt.plot([result['test_acc'] for result in low_precision_history], label='test')
        plt.legend()
        plt.title(f'{bit}-BIT precision, D={dim}')
        plt.xlabel('epochs')
        plt.ylabel('accuracy')
        plt.show()
```
